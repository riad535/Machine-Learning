```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
avg = df.groupby(by='variety',as_index=False).mean()
bar_df = avg.drop('variety',axis=1)
bar_df.head()
```

```python
bar_df.plot(kind="bar",figsize=(10,5))
plt.title("Bar Chart")
plt.xlabel("Targets")
plt.ylabel("Measurements")
plt.show()
```
